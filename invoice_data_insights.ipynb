We import the pickle file, and collect it using pandas DataFrame.

```python
import pickle
import pandas as pd
dbfile = open("df_train_api.pk", 'rb')
df = pickle.load(dbfile)
```

We set necessary height and width in order to display the rows untruncated.

```python
pd.set_option('display.height', 500)
pd.set_option('display.max_rows', 500)
```

```python
df
```

As the column 'data' represents different files here we group the data according to it.

```python
df.groupby('data').groups
```

Getting no. of unique values in the column 'data'.

```python
data = df['data'].unique().tolist()
```

Iterating the list to print the next row to Invoice No. to get the invoice no. Not in all the cases, the invoice no. is located 
in the row next to the one having entry 'Invoice No.' Same applies to the other features we are after.

```python
for i in data:
    index = df.index[(df['data']==i) & (df['groups']=='Invoice No.')].tolist()[0] + 1
    df.loc[[index]]
```

Now, getting the dates.

```python
for i in data:
        print(df[(df['data']==i) & (df['groups']=='Dated')])
```

We find the data for the rows having GST information.

```python
import re
df[df.groups.str.contains('GSTIN', flags=re.IGNORECASE, regex=True)]
```

We find the data for the rows having amount information.

```python
import re
df[df.groups.str.contains('amount', flags=re.IGNORECASE, regex=True)]
```
